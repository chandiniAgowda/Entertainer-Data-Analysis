```python
# import pandas and numpy library
import pandas as pd
import numpy as np

```

```python
#reading the dataset "Entertainer_data.csv" and storing it in a dataframe
#using pd.read_csv() function
df=pd.read_csv('/content/Entertainer_data.csv')
```

```python
#displaying top 5 rows of the dataset
df.head(5)
```

```python
#checking the datatypes of all the columns
df.info()
```

```python
#change Year of Breakthrough/#1 Hit/Award Nomination column name as Year of Breakthrough and Gender (traditional) as Gender
df.rename(columns={'Year of Breakthrough/#1 Hit/Award Nomination':'Year of Breakthrough','Gender (traditional)':'Gender'},inplace=True)
#displaying top 5 rows after changing the column names
df.head(5)
```

```python
# check null values

df.isnull().sum()
```

```python
#Column 'Year of first Oscar/Grammy/Emmy has some blank spaces. Relaced them with NA

df['Year of First Oscar/Grammy/Emmy'].fillna('NA', inplace=True)
```

```python
#Column 'Year of Death' has some blank spaces. Relaced them with a placeholder 'Still Alive'
df['Year of Death'].fillna('Still Alive', inplace=True)
```

```python
# displaying top 5 row
df.head(5)
```

```python
#checking null values after replacing them

df.isnull().sum()
```

```python
#Created a new column 'Age at Breakthrough' which is 'Year of Breakthrough' subtracting 'Birth Year'.

df['Age at Breakthrough']=df['Year of Breakthrough']-df['Birth Year']
```

```python
#Created another column 'Active Years' which is 'Year of Last Major Work' subtracting 'Year of Breakthrough'

df['Active Years']=df['Year of Last Major Work (arguable)']-df['Year of Breakthrough']

#displaying top 5 rows of the dataset
df.head(5)
```

```python
#

df.to_csv('Entertainer_cleaned_data.csv', index=False)
```
